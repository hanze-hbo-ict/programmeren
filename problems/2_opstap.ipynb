## Debuggen

Het kunnen vinden van bugs in conditionele statements.

### Opgave

Tijdens een busreis worden studenten verdeeld over verschillende bussen aan de hand van hun achternaam.

| Bus | Eerste letter achternaam |
| --- | ------------------------ |
| 1   | A t/m E                  |
| 2   | F t/m J                  |
| 3   | K t/m O                  |
| 4   | Q t/m U                  |
| 5   | V t/m Z                  |

Bijvoorbeeld, een student met de achternaam "NiÃ«l" komt in bus 3 en een student met de achternaam "Hoebe" komt in bus 2

Gegeven is een variabele `naam` die staat voor de achternaam (zonder voorvoegsels en met hoofdletter) van een student. De opdracht is om een conditionele statement te schrijven om de juiste bus nummer af te drukken.

````python
naam = "Hoebe"
````

Deze opdracht is op meerdere manieren op te lossen en hieronder staan een aantal uitwerkingen, maar er zit telkens een fout in!

**Probeer bij elke uitwerking te bepalen welke fout wordt gemaakt waardoor de code niet correct is.**

#### Uitwerking 3-A

````python
if naam[0] <= 'E':
    print(1)
if naam[0] <= 'J':
    print(2)
if naam[0] <= 'O':
    print(3)
if naam[0] <= 'U':
    print(4)
else
    print(5)
````

#### Uitwerking 3-B

````python
if naam[0] >= 'E':
    print(1)
elif naam[0] >= 'J':
    print(2)
elif naam[0] >= 'O':
    print(3)
elif naam[0] >= 'U':
    print(4)
else:
    print(5)
````

#### Uitwerking 3-C

````python
if naam[0] <= 'E':
    print(1)
else:
    if naam[0] <= 'J':
        print(2)
    elif naam[0] <= 'O':
        print(3)
        if naam[0] <= 'U':
            print(4)
        else:
            print(5)
````

#### Uitwerking 3-D

````python
if naam[0] <= 'E':
    print(1)
elif naam[0] <= 'J':
    print(2)
    elif naam[0] <= 'O':
        print(3)
    elif naam[0] <= 'U':
        print(4)
    else:
        print(5)
````

#### Uitwerking 3-E

````python
if naam[0] <= 'Z':
    print(5)
elif naam[0] <= 'U':
    print(4)
elif naam[0] <= 'O':
    print(3)
elif naam[0] <= 'J':
    print(2)
else:
    print(1)
````

#### Uitwerking 3-F

````python
if naam <= 'E':
    print(1)
elif naam <= 'J':
    print(2)
elif naam <= 'O':
    print(3)
elif naam <= 'U':
    print(4)
else:
    print(5)
````

#### Uitwerking 3-G

````python
if naam < 'F'
    print(1)
elif naam < 'K'
    print(2)
elif naam < 'P'
    print(3)
elif naam < 'V'
    print(4)
else:
    print(5)
````
