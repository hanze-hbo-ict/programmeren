### Faculteit met `for`

````python
def fac(n):
    result = 1                      # not the result yet!

    for number in range(1, n + 1):  # range start at 0, add one!
        result *= number            # result = result * i

    return result                   # return the result
````

````python
fac(5)
````

## Twee typen `for`

Elementen versus index

### Op basis van element 

```python
my_list = [3, 15, 17, 7]

for item in my_list:
    print(item)
```

```python
string = "een fijne lus"

for char in string:
    print(char)
```

### Op basis van index

```python
my_list = [3, 15, 17, 7]

for ix in range(len(my_list)):
    print(my_list[ix])
```

```python
string = "een fijne lus"

for ix in range(len(string)):
    print(string[ix])
```

Let op, het is niet heel gewoon om in lussen te printen, maar is wel heel nuttig voor debuggen!

### Welke van de twee?

Elementen: eenvoudiger

Indices: flexibeler


Denk aan het "time to think this over! " voorbeeld, in de lus kon op basis van de index steeds "terug" worden gekeken in de string met `string[ix - 1]`!

## `while` lussen

Een lus tot een bepaalde **conditie** is bereikt.

Tests?

- `42 == 42`
- `guess > 42`
- ...

### Ontsnappen

````python
import random

guess = 0  # starting value, not the final or desired value!

while guess != 42:  # test to see if we keep looping
    print('Help! Let me out!')
    guess = random.choice([41, 42, 43])  # watch out for infinite loops!

print('At last!')  # after the loop ends
````

## `continue` en `break`

De instructie `break` zal een lus meteen laten stoppen en het programma gaat verder onder de lus.

De instructie `continue` gaat naar het einde van de lus en indien nodig wordt de lus herhaald.
