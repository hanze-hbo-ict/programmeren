$$
\begin{aligned}
222_4 &= 2 \cdot 16 + 2 \cdot 4 + 2 \cdot 1 &= 42 \\
60_7 &= 6 \cdot 42 + 0 \cdot 1 &= 42 \\
46_9 &= 4 \cdot 9 + 6 \cdot 1 &= 42 \\
39_{11} &= 3 \cdot 11 + 9 \cdot 1 &= 42 \\
\end{aligned}
$$
