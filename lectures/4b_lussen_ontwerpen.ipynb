# Lussen ontwerpen. 

Zodra je een probleem wil oplossen met lussen moet je een paar belangrijke stappen maken. 

1. **Defineer de start waarde als `resultaat`**. Dit is afhankelijk van wat de functie teruggeeft en is vaak een keuze tussen een lijst, string of integer. Als de functie bijvoorbeeld een lijst teruggeeft, dan is de start waarde ook een lijst. Naast het beslissen welk datatype het is, moet ook de start waarde bepaald worden. Stel jezelf de vraag wat de waarde is in het begin. Als je bijvoorbeeld wil tellen hoeveel elementen er in een lijst zitten, begin je te tellen vanaf 0. De start waarde is dan ook 0. Waar je goed op moet letten is met vermenigvuldigen. Dan is de start waarde nooit gelijk aan 0, want als je vermenigvuldigt met 0, blijft het 0. 


2. **CreÃ«er de lus.** Als je vooraf niet weet hoe vaak de lus herhaald moet worden om een bepaald doel te bereiken, dan gebruik je een while-lus. Belangrijk is dan om goed na te denken over de voorwaarde zodat de lus op het juist moment stopt, of gebruik een break om de lus te stoppen. In andere situaties kan je een for-lus gebruiken. Het is dan nog goed om erover na te denken welk type for-lus. Het lopen over alle elementen heen (`for x in l`) is zeer handig voor als je alle character in een string langs moet gaan, of de elementen van een lijst moet langslopen. De andere type is met range (`for i in range(n)`) waarbij `n` aangeeft hoe vaak de lus moet herhalen. Deze methode is vooral handig voor problemen waarbij het bijhouden van de index belangrijk is of wanneer je meer dan 1 element uit de lijst wilt gebruiken. Met `l[i] < l[i-1]`bijvoorbeeld kan je twee elementen in een lijst vergelijken met elkaar.

3. **Defineer de voorwaarde indien nodig.?** Bepaal aan welke voorwaarde voldaan moet worden om het resultaat up te daten. Als je bijvoorbeeld wil tellen hoeveel even getallen er in een lijst zitten, is de voorwaarde dus dat het element een even getal moet zijn. 


4. **Update `resultaat`** Als aan de juiste voorwaarde wordt voldaan, dan kunnen we het resultaat updaten. We kunnen er bijvoorbeeld 1 bij optellen, of een letter toevoegen aan ons resultaat. 


5. **Stop de lus indien nodig.** Deze is zeker belangrijk bij while-lussen. Als het gewenste resutlaat is bereikt, dan kan de lus voortijdig gestopt worden met `break`. In sommige gevallen kan je een lus ook stoppen met een return statement als deze lus onderdeel is van een functie. De return statement beindigd de hele functie namelijk. 

6. **Geef het `resultaat` terug aan het einde van de lus** Zodra de lus helemaal klaar is, kan het resultaat teruggegeven worden met een return statement. Belangrijk is om op de indent te letten zodat de return niet per ongeluk in de lus staat en dan vroegtijdig de functie beindigt. 

## Voorbeelden

```python
def count(s):
    """
    Ontvang een string, telt hoeveel characters er in de string zit en geeft het resultaat terug als integer.  
    """
    result = 0                 # Defineer start waarde. Tellen vanaf 0. We hebben nog 0 characters geteld. 

    for c in s:                # Creeer de lus.Loop door alle elementen heen. 
        result = result + 1    # Update resultaat. Tel 1 erbij op. 

    return result              # einde lus; geef het resultaat terug. 

print(count("Hanze"))
```

```python
def klinkers(s):
    """
    Ontvang een string en geeft een string terug met enkel de klinkers.
    """
    result = ''                     # Defineer de start waarde. Functie geeft een string terug, dus begin met een lege string. We hebben nog geen klinkers gevonden

    for c in s:                     # creeer de lus. Loop door alle elementen heen. 
        if c in "eaoiu":            # Defineer de voorwaarde. Als c een klinker is...
            result = result+ c      # ...Voeg de klinker toe aan het resultaat

    return result                   # einde lus; geef het resultaat terug. 

print(klinkers("Hanze"))
```

```python
def bevat(s, c):
    """
    Ontvang string s en character c. Geeft True terug als  c in s zit. 
    """
    result = False              # Defineer de start waarde. Funtie geeft een boolean terug, dus defineer False. We hebben c nog niet gevonden. 

    for x in s:                 # creeer de lus. Loop door alle elementen heen. 
        if x  ==  c :           # Defineer de voorwaarde. als het element gelijk is aan c...
            result = True       # ... update het resultaat naar True. 
            break               # Stop de lus. Eind resultaat is gevonden

    return result               # einde lus; geef het resultaat terug. 

print(bevat("Hanze", "n"))
```

```python
def index_of(s, c):
    """
    Ontvang string s en character c. Geet de index terug van de character c als een integer. Als c niet in de string zit, geef -1 terug. 
    """
    result = False              # Defineer de start waarde. Funtie geeft een integer terug. In dit geval is de start waarde -1. Het element is aan het begin nog niet gevonden. 

    for i in range(len(s)):     # creeer de lus. Gezien we de index zoeken van c, gebruiken we de index type lus. 
        if s[i]  ==  c :        # Defineer de voorwaarde. als het element s[i] gelijk is aan c...
            result = i          # ... resultaat is de huidige index. 
            break               # Stop de lus. Eind resultaat is gevonden

    return result               # einde lus; geef het resultaat terug. 

print(index_of("Hanze", "n"))
```

## Optelling: Som(l)

De functie `som(l)` ontvang een lijst `l` met integers en telt alle integers bij elkaar op en geeft het resultaat terug als een integer. 

### plan. 
1. Defineer de start waarde als `resultaat`: 
De functie geeft een integer terug na het optellen van een aantal getallen. We beginnen met tellen vanaf 0. 
`result = 0 `

2. Creeer de lus: 
We willen door alle elementen van de lijst lopen. 
`for e in l`

3. Defineer de voorwaarde indien nodig: 
Er is geen voorwaarde waar het element aan moet voldoen. 

4. Update `resultaat`: 
Het element moet bij het resultaat opgeteld worden.
`resultaat = resultaat + c`

5. Stop de lus indien nodig:
De lus hoeft niet vroeg tijdig gestopt te worden gezien we alle getallen langs moeten om ze bij elkaar op te tellen. 

6. Geef het `resultaat` terug aan het einde van de lus: 
`return resultaat`

```python
def som(s):
    """
    Ontvang een lijst met integers, telt alle integers bij elkaar op en geeft het resultaat terug als integer.  
    """
    result = 0                 # Defineer start waarde. Tellen vanaf 0

    for e in s:                # Creeer de lus.Loop door alle elementen heen. 
        result = result + e   # Update resultaat. Tel e erbij op. 

    return result              # einde lus; geef het resultaat terug. 

print(som([1,2,3,4,5]))
```
