# Fijne functies

```python
def sq(x):
    """Returns Square of the argument

    :param x: The value to square
    :type x: int, float
    :rtype: int, float
    """
    return x * x
```

```python
def interp(low, hi, fraction):
    """Calulcates fraction

    :param low: lowest value
    :param hi: highest value
    :param fraction: fraction 
    :type low, hi, fraction: int, float
    :rtype: float
    """

    return low + (hi - low) * fraction 
```

```python
def checkends(string):
    return string[0] == string[-1] 
```

```python
def flipside(string):
    halfLength = int(len(string)/2)
    return string[halfLength :] + string[:halfLength]
```

```python
def convert_from_seconds(seconds_in):
    days = int(seconds_in/86400)
    left = seconds_in % 86400
    hours = int(left/3600)
    left = left % 3600
    minutes = int(left/60)
    seconds = left % 60
    return [days, hours, minutes, seconds]
```
